```python
age = datas['age'].value_counts()
sex = datas['sex'].value_counts()
children = datas['children'].value_counts()
smoker = datas['smoker'].value_counts()
region = datas['region'].value_counts()



label_age = age.index
size_age = age.values

label_sex = sex.index
size_sex = sex.values

label_children = children.index
size_children = children.values

label_smoker = smoker.index
size_smoker = smoker.values

label_region = region.index
size_region = region.values

colors = []
trace = go.Pie(
         labels = label_age, values = size_age, marker = dict(colors = colors), name = 'Age', hole = 0.3)

colors2 = ['aqua', 'gold']
trace2 = go.Pie(labels = label_sex, values = size_sex, marker = dict(colors = colors2), name = 'Sex',
                hole = 0.3)

colors3 = []
trace3 = go.Pie(
         labels = label_children, values = size_children, marker = dict(colors = colors), name = 'Children', hole = 0.3)

colors4 = []
trace4 = go.Pie(
         labels = label_smoker, values = size_smoker, marker = dict(colors = colors), name = 'Smoker', hole = 0.3)

colors5 = []
trace5 = go.Pie(
         labels = label_region, values = size_region, marker = dict(colors = colors), name = 'Region', hole = 0.3)

df = [trace]
df2 = [trace2]
df3 = [trace3]
df4 = [trace4]
df5 = [trace5]

layout1 = go.Layout(
           title = 'Age')
layout2 = go.Layout(
           title = 'Sex')
layout3 = go.Layout(
           title = 'Children')
layout4 = go.Layout(
           title = 'Smoker')
layout5 = go.Layout(
           title = 'Region')

fig = go.Figure(data = df, layout = layout1)
fig2 = go.Figure(data = df2, layout = layout2)
fig3 = go.Figure(data = df3, layout = layout3)
fig4 = go.Figure(data = df4, layout = layout4)
fig5 = go.Figure(data = df5, layout = layout5)


py.iplot(fig)
py.iplot(fig2)
py.iplot(fig3)
py.iplot(fig4)
py.iplot(fig5)
```

```python
factors=['age', 'sex', 'bmi', 'children', 'expenses']
data.groupby('sex')[factors].mean()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data.sex = le.fit_transform(data.sex)
data.smoker = le.fit_transform(data.smoker)
data.region = le.fit_transform(data.region)
```

```python
x=data.drop(['expenses'],axis=1)
x
```

```python
y=data.expenses
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
print("Training Accuracy :", model.score(X_train, y_train))
print("Testing Accuracy :", model.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model1 = DecisionTreeRegressor()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
y_pred
```

```python
print("Training Accuracy :", model1.score(X_train, y_train))
print("Testing Accuracy :", model1.score(X_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
model2 = RandomForestRegressor()
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
y_pred
```

```python
print("Training Accuracy :", model2.score(X_train, y_train))
print("Testing Accuracy :", model2.score(X_test, y_test))
```
